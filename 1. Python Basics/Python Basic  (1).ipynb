Question 2: Explain the difference between List and Tuple in Python.

Answer:
Lists and tuples are both data structures in Python that are used to store multiple elements. However, they have several differences.

1. Mutability:
Lists are mutable, which means their elements can be changed after creation.
Tuples are immutable, which means their elements cannot be changed once created.

2. Syntax:
Lists are created using square brackets [ ].
Tuples are created using round brackets ( ).

3. Performance:
Tuples are faster than lists because they are immutable.

4. Usage:
Lists are used when data needs to be modified.
Tuples are used when data should remain constant and unchanged.

5. Example:
List Example: [1, 2, 3]
Tuple Example: (1, 2, 3)

In conclusion, lists are flexible while tuples are more secure and faster.

Question 3: What is a function in Python? Why are functions useful?

Answer:
A function in Python is a block of code that performs a specific task. Functions help in dividing a large program into smaller and manageable parts.

Functions are defined using the 'def' keyword followed by the function name and parentheses.

Functions are useful for the following reasons:

1. Code Reusability:
Functions allow the same code to be used multiple times without rewriting it.

2. Reduces Complexity:
Functions break a large program into smaller parts, making it easier to understand.

3. Saves Time:
It reduces the need to write repetitive code.

4. Easy Debugging:
Errors can be easily found and fixed in functions.

5. Better Organization:
Functions help in organizing the code properly.

Thus, functions play an important role in making programs efficient and manageable.

```python
# Question 5: Program to check whether a number is even or odd

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python
Explanation:
This program takes a number as input from the user. It checks whether the number is divisible by 2 using the modulus
operator (%). If the remainder is 0, the number is even; otherwise, it is odd.
```

```python
# Question 8: Remove duplicate values from a list

numbers = [1, 2, 2, 3, 4, 4, 5]

unique_numbers = list(set(numbers))

print("List after removing duplicates:", unique_numbers)
```

```python
Explanation:
This program removes duplicate values from a list using a set. 
A set does not allow duplicate elements, so it automatically removes repeated values.
```

```python
# Question 9: Function to return square of a number

def square(num):
    return num * num

result = square(5)

print("Square is:", result)
```

Explanation:
This program defines a function named square() which takes a number as input and returns its square. 
The function is then called with the value 5, and the result is printed.

```python
# Question 10: Count occurrences of a number in a list

numbers = [2, 3, 4, 2, 5, 2]

count = numbers.count(2)

print("2 appears", count, "times in the list")
```

Explanation:
This program counts how many times a number appears in a list using the count() function.
In this example, the number 2 appears three times.
