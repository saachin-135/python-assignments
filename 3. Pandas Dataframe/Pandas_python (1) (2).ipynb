Q1. Creating a DataFrame

This dataset will be useful for performing further analysis such as filtering, aggregation, and statistical operations.

Q4. Indexing using .loc


Display row with index 10 and Sales value at index 20.

```python
df.loc[10]
df.loc[20, "Sales"]
```

Explanation

.loc is used for label-based indexing.
It allows analysts to access specific rows and columns using actual labels.

This is useful when:

You know exact row index
You want precise data extraction

Q5. Indexing using .iloc


Display first 6 rows & first 4 columns and Sales of 3rd row.

```python
df.iloc[0:6, 0:4]
df.iloc[2]["Sales"]
```

Explanation

.iloc is position-based indexing.
It is useful when working with numerical positions instead of labels.

This method is commonly used in data slicing and feature extraction.

Q7. Filtering Data

Q8. Multiple Condition Filtering


Filter rows where:

Region = West
Sales > 1200

```python
df[(df["Region"] == "West") & (df["Sales"] > 1200)]
```

Explanation

Multiple condition filtering helps refine analysis further.
Using logical operators (&), we can apply more than one condition simultaneously.

This is useful for targeted insights, such as identifying high sales in a specific region.
