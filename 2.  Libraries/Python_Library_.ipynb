### Question 1: Load the dataset and display basic information

Explanation:
In this question, the dataset is loaded using the pandas library. The head() function is used to display the first five rows of the dataset, and the info() function provides details about column names, data types, and missing values.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/retail_dataset.csv")

df.head()
df.info()
```

### Question 2: Inspect the dataset and identify missing values

Explanation:
This step is used to check how many missing (null) values are present in each column of the dataset. It helps in understanding data quality.

```python
df.isnull().sum()
```

### Question 3: Replace missing values in Quantity and Price

Explanation:
Missing values in the Quantity and Price columns are replaced with their respective mean values. This ensures that the dataset remains consistent.

```python
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
df['Price'].fillna(df['Price'].mean(), inplace=True)
```

### Question 4: Remove rows with missing values

Explanation:
Rows that contain missing values in Product Category or Region columns are removed to improve data accuracy.

```python
df.dropna(subset=['Product Category', 'Region'], inplace=True)
```

### Question 5: Create a new column Revenue

Explanation:
A new column called Revenue is created by multiplying Quantity and Price. It represents total earnings from each transaction.

```python
df['Revenue'] = df['Quantity'] * df['Price']
df.head()
```

### Question 6: Calculate total revenue

Explanation:
NumPy is used to calculate the total revenue by summing all values in the Revenue column.

```python
import numpy as np

total_revenue = np.sum(df['Revenue'])
print("Total Revenue:", total_revenue)
```

### Question 7: Revenue by product category

Explanation:
The dataset is grouped by Product Category, and the total revenue for each category is calculated.

```python
category_revenue = df.groupby('Product Category')['Revenue'].sum()
print(category_revenue)
```

### Question 8: Top and Bottom categories

Explanation:
The top 3 categories with highest revenue and bottom 3 categories with lowest revenue are identified.

```python
top3 = category_revenue.sort_values(ascending=False).head(3)
bottom3 = category_revenue.sort_values().head(3)

print("Top 3 Categories:\n", top3)
print("Bottom 3 Categories:\n", bottom3)
```

### Question 9: Revenue by region

Explanation:
The dataset is grouped by Region to calculate total revenue in each region. The highest and lowest revenue regions are identified.

```python
region_revenue = df.groupby('Region')['Revenue'].sum()

print(region_revenue)
print("Highest Revenue Region:", region_revenue.idxmax())
print("Lowest Revenue Region:", region_revenue.idxmin())
```

### Question 10: Monthly revenue and statistics

Explanation:
The Date column is converted into datetime format to extract the month. Monthly revenue is calculated. Additionally, mean, median, and standard deviation of Revenue are computed using NumPy.

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

df['Month'] = df['Date'].dt.month

monthly_revenue = df.groupby('Month')['Revenue'].sum()

print(monthly_revenue)
```
